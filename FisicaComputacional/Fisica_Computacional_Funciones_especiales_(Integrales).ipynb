## **Funciones de Fresnel.**

```python
import numpy as np
import matplotlib.pyplot as plt
def simpsion(f,a,b,N):
  h=abs(b-a)/N
  s_p,s_i=0,0
  fa=f(a)
  fb=f(b)
  for k in range(2,N-1,2):#suma de pares.
    s_p+=f(a+k*h)
  for j in range(1,N,2): #suma de impares.
    s_i+=f(a+j*h)
  s=(h/3)*(fa+fb+4*s_p+2*s_i)
  return s
```

```python
def S(x, N=1000):
  if x > 0:
    return simpsion(lambda t: np.sin( t**2), 0, x, N)
  else:
    return -simpsion(lambda t: np.sin( t**2), 0, x, N)
x_val = np.linspace(-10, 10, 500)
S_val = [S(x) for x in x_val]
plt.plot(x_val, S_val)
plt.xlabel('x')
plt.ylabel('S(x)')
plt.grid(True)
plt.show()
```

```python
def C(x, N=1000):
    if x > 0:
      return simpsion(lambda t: np.cos( t**2), 0, x, N)
    else:
      return -simpsion(lambda t: np.cos( t**2), 0, x, N)
C_val = [C(x) for x in x_val]
plt.plot(x_val, C_val)
plt.xlabel('x')
plt.ylabel('C(x)')
plt.grid(True)
plt.show()
```

```python
plt.plot(C_val, S_val)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('C(x)')
plt.ylabel('S(x)')
plt.grid(True)
plt.show()
```

# Funciones de **Bessel**

```python
n=[0,1,2,3,4,5]
def J(x,n,N=1000):
    return simpsion(lambda t: np.cos(n*t-x*np.sin(t)), 0, np.pi, N)
x_val = np.linspace(0, 20, 500)
J_val = [[J(x, ni) for x in x_val] for ni in n]
for i in range(0,5):
  plt.plot(x_val, J_val[i], label=f'n={i}')
plt.xlabel('x')
plt.ylabel('J(x)')
plt.grid(True)
plt.show()
```

# Polinomios de **Legendre.**

```python
n=[0,1,2,3,4,5]
def L(x,n,N=1000):
    return simpsion(lambda t: (x+np.cos(t)*np.emath.sqrt(x**2-1))**n, 0, 2*np.pi, N)
x_val = np.linspace(-1, 1, 500)
L_val = [[L(x, ni) for x in x_val] for ni in n]
for i in range(0,5):
  plt.plot(x_val, L_val[i], label=f'n={i}')
plt.xlabel('x')
plt.ylabel('L(x)')
plt.grid(True)
plt.show()
```
