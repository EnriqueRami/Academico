# **Caminante aleatorio 1D**

```python
import random
import matplotlib.pyplot as plt
import numpy as np

#Primeramente realizamos el ejercicio caminante aleatorio en 1 dimensión.
def caminante_aleatorio_1D(x0,N): #  x0: Posición inicial del caminante (cambia el punto de partida del caminante (no afecta el comportamiento general)) y N: Número de pasos que dará el caminante (Afecta cuán lejos puede llegar el caminante (a mayor N, mayor dispersión posible))
  for i in range(N):
    x0 += random.choice([-1,1]) # En cada paso, elige aleatoriamente entre -1 (izquierda) o +1 (derecha).
  return x0 #Retorna (regresa) la posición final después de N pasos.

#Posteriormente, simulamos múltiples caminantes aleatorios independientes.
def M_caminantes():
  camino=[] #Iniciamos una lista vacía que va a contener, las posiciones finales de 1000 caminantes después de 100 pasos cada uno.
  for i in range(1000): #(1000): Número de caminantes (aumentarlo da mejores estadísticas pero más tiempo de cálculo).
    camino.append(caminante_aleatorio_1D(0,100)) #Guardamos los valores del camino que va tomando en la lista. El (100) es el número de pasos por caminante (similar a N en caminante_aleatorio_1D).
  return camino #Retorna (regresa) la lista con las posiciones finales de 1000 caminantes después de 100 pasos cada uno.
```

```python
print(M_caminantes()) #Muestra la lista completa de las posiciones finales de los 1000 caminantes, cada valor representa dónde terminó un caminante después de 100 pasos aleatorios (empezando en 0).
print(len(M_caminantes())) #Imprime 1000, porque la función genera 1000 caminantes (valor fijo en el range(1000)).
```

```python
x=np.linspace(-30, 30,100) #Creamos un array de 100 puntos espaciados uniformemente entre -30 y 30. Si modificamos (-30/30 cambiamos el rango visualizado (debe cubrir las posiciones de los caminantes) y si cambiamos (100) afectamos la suavidad de la curva gaussiana (más puntos = más suave)
sigma = 100 #Definimos la varianza (σ²) de la distribución gaussiana. Si aumentamos sigma la campana gaussiana será más ancha y se perderá precisión teórica.
gaussiana= lambda x: (1/(np.sqrt(2*np.pi)))*np.exp(-x**2/sigma)*210 #Definimos la funcion lambda que define una gaussiana. El 210 es un factor empírico de escala.

plt.hist(M_caminantes(),bins=35, label='Caminante aleatorio', color='skyblue') #Generamos el histograma de las posiciones finales de los caminantes ((bins=50): Número de divisiones del histograma, demasiados bins para 1000 muestras puede verse "ruidoso")
plt.plot(x, gaussiana(x), 'ro-', label='Distribución sub-Poissoniana') #Superponemos la curva gaussiana.
plt.grid() #Añademos una cuadrícula al gráfico para mejor lectura :)
plt.title("Caminante aleatorio 1D") #Título del gráfico.
plt.xlabel("Posición") #Etiqueta del eje X.
plt.ylabel("Frecuencia") #Etiqueta del eje Y.
plt.show() #Mostramos el gráfico.
```

# **Caminante aleatorio 2D**

```python
import random
import matplotlib.pyplot as plt
import numpy as np


#Ahora realizamos el caso del caminante aleatorio en 2D, pero en este caso consideramos la opcion si el caminante se queda quieto.
def caminante_aleatorio_1Dx(x0,N): #x0: Posición inicial del caminante para el eje x y N: Número de pasos que dará el caminante.
  for i in range(N):
    x0 += random.choice([-1,0,1])  #Paso aleatorio: izquierda, quieto o derecha.
  return x0 #Retorna (regresa) la posición final (en x) después de N pasos.

def caminante_aleatorio_1Dy(y0,N): #De igual manera realizamos el codigo para un caminante aleatorio en el eje Y ..
  for i in range(N):
    y0 += random.choice([-1,0,1]) #Paso aleatorio: izquierda, quieto o derecha.
  return y0 #Retorna (regresa) la posición final (en y) después de N pasos.

def M_caminantesx(): #De la misma manera que el caso en 1D generamos 1000 caminantes en el eje X con 1000 pasos cada uno.
  camino_x=[] #Iniciamos una lista vacía que va a contener, las posiciones finales en X de 1000 caminantes después de 100 pasos cada uno.
  for i in range(1000): #(1000): Número de caminantes (aumentarlo da mejores estadísticas pero más tiempo de cálculo).
    camino_x.append(caminante_aleatorio_1Dx(0,1000)) #Guardamos los valores del camino que va tomando en la lista.
  return camino_x

def M_caminantesy(): #Generamos 1000 caminantes en el eje Y con 1000 pasos cada uno.
  camino_y=[] #Iniciamos una lista vacía que va a contener, las posiciones finales en Y de 1000 caminantes después de 100 pasos cada uno.
  for i in range(1000): #(1000): Número de caminantes (aumentarlo da mejores estadísticas pero más tiempo de cálculo)
    camino_y.append(caminante_aleatorio_1Dy(0,1000)) #Guardamos los valores del camino que va tomando en la lista.
  return camino_y
```

```python
print(M_caminantesx()) #Muestra la lista completa de las posiciones finales en eje X de los 1000 caminantes, cada valor representa dónde terminó un caminante después de 100 pasos aleatorios (empezando en 0).
print(M_caminantesy()) #Muestra la lista completa de las posiciones finales en el eje Y de los 1000 caminantes, cada valor representa dónde terminó un caminante después de 100 pasos aleatorios (empezando en 0).
```

```python
plt.hist(M_caminantesx(), bins=35, label='Caminante aleatorio', color='skyblue') #Generamos el histograma de las posiciones finales de los caminantes en el eje X, con bins=35 (número de divisiones del histograma).
plt.grid() #Añadimos una cuadrícula al gráfico para más placer :)
plt.title("Caminante aleatorio 1D") #Título del gráfico.
plt.xlabel("Posición") #Etiqueta del eje X.
plt.ylabel("Frecuencia") #Etiqueta del eje Y.
plt.show() #Mostramos el gráfico.
```

```python
plt.hist(M_caminantesy(), bins=35, label='Caminante aleatorio', color='green') #Generamos el histograma de las posiciones finales de los caminantes en el eje Y, con bins=35 (número de divisiones del histograma).
plt.grid() #Añadimos una cuadrícula al gráfico para mejor visualización :)
plt.title("Caminante aleatorio 1D") #Título del gráfico.
plt.xlabel("Posición") #Etiqueta del eje X.
plt.ylabel("Frecuencia")#Etiqueta del eje Y.
plt.show() #Mostramos el gráfico.
```

```python
#Gráficamos la distribucion 2D de caminantes aletorios.
plt.figure(figsize=(10, 8)) #Generamos una figura de 10 x 8.
plt.hist2d(M_caminantesx(),M_caminantesy(), bins=35, label='Caminante aleatorio')  #Generamos el histograma en 2D, con bins=35 (número de divisiones del histograma).
plt.scatter([0], [0], c='red', s=50, label='Origen') # Marcamos el origen
plt.title('Distribución 2D de Caminantes Aleatorios.') #Título del gráfico.
plt.xlabel('Posición final en X') #Etiqueta del eje X.
plt.legend() #Muestra la leyenda del grafico.
plt.ylabel('Posición final en Y') #Etiqueta del eje Y.
plt.grid(alpha=0.3)  # Grid semitransparente
```

```python
#Primero creamos la malla de coordenadas.
x=np.linspace(-120, 120,1000) #Creamos un array (para el eje X) de 100 puntos espaciados uniformemente entre -120 y 120.
y=np.linspace(-120, 120,1000) #De igual manera creamos (para el eje Y) un array de 100 puntos espaciados uniformemente entre -120 y 120.
x,y= np.meshgrid(x,y) #Generamos una malla de coordenadas 2D.
N=1/5000   # Parámetro de dispersión.
r=np.sqrt(x**2+y**2) # Calculamos las distancias radiales.
f = lambda r,N: (1/2*N)*np.exp(-r**2/2*N) # Distribución gaussiana 2D normalizada.
fig, ax = plt.subplots() #Creamos una figura y ejes.
ax.contour(x,y,f(r,N), cmap = 'cool') #Dibujamos las líneas de contorno.
plt.title('Distribución Gaussiana 2D.') #Título del gráfico.
plt.xlabel('Coordenada X', fontsize=12) #Etiqueta del eje X.
plt.ylabel('Coordenada Y', fontsize=12) #Etiqueta del eje Y.
plt.grid(alpha=0.2)  # Cuadrícula sutil
plt.tight_layout() # Ajuste automático de márgenes
plt.show() # Muestra el gráfico
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de Simpson para calcular la integral
def simpson_integral(f, a, b, N):
    if N % 2 == 1:  # N debe ser par para Simpson
        N += 1
    h = (b - a) / N
    x = np.linspace(a, b, N + 1)
    fx = f(x)
    integral = (h / 3) * (fx[0] + 4 * np.sum(fx[1:-1:2]) + 2 * np.sum(fx[2:-2:2]) + fx[-1])
    return integral

# Función de Bessel J_m(x)
def bessel_function(m, x, N=1000):
    def integrand(theta):
        return np.cos(m * theta - x * np.sin(theta))
    return (1 / np.pi) * simpson_integral(integrand, 0, np.pi, N)

# Generar los valores de J_0, J_1 y J_2
x_values = np.linspace(0, 20, 500)
J0 = [bessel_function(0, x) for x in x_values]
J1 = [bessel_function(1, x) for x in x_values]
J2 = [bessel_function(2, x) for x in x_values]

# Graficar las funciones de Bessel
plt.figure(figsize=(10, 6))
plt.plot(x_values, J0, label=r"$J_0(x)$")
plt.plot(x_values, J1, label=r"$J_1(x)$")
plt.plot(x_values, J2, label=r"$J_2(x)$")
plt.title("Funciones de Bessel $J_0$, $J_1$ y $J_2$")
plt.xlabel("x")
plt.ylabel(r"$J_m(x)$")
plt.legend()
plt.grid(True)
plt.show()
```
