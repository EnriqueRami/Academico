# **Metodo de Euler**

```python
import numpy as np
import matplotlib.pyplot as plt

f= lambda x,t: -x**3+np.sin(t)
a,b=0.0,10.0
N=60
h=(b-a)/N
x=0.0
puntos_t=[]
puntos_x=[]
for t in np.arange(a,b,h):
    puntos_x.append(x)
    x=x+h*f(x,t)
    puntos_t.append(t)
```

```python
plt.plot(puntos_t,puntos_x)
plt.grid()
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Metodo de Euler')
plt.show()
```

# **Ley de enfriamiento de Newton**

```python
def Temp(t):
  T_amb=23
  k=1/2
  T0=100
  return T_amb + (T0-T_amb)*np.exp(-k*t)

puntos_T=[]
for t in np.arange(0,10,0.1):
  puntos_T.append(Temp(t))
```

```python
T_amb=23
k=1/2
g= lambda T: -k*(T-T_amb)
a,b=0.0,80.0
N=200
h=(b-a)/N
T=100.0
puntos_T=[]
puntos_t=[]
for t in np.arange(a,b,h):
    puntos_T.append(T)
    T=T+h*g(T)
    puntos_t.append(t)
```

```python
plt.plot(puntos_t, puntos_T)
plt.grid()
plt.xlabel('t')
plt.ylabel('T(t)')
plt.title('Ley de enfriamiento de Newton (numerica)')
plt.show()
```

# **Metodo de Runge-kutta**

```python
import numpy as np
import matplotlib.pyplot as plt
f= lambda x,t: -x**3+np.sin(t)
a,b=0.0,10.0
N=100
h=(b-a)/N
x=0.0
puntos_t=[]
puntos_x=[]
for t in np.arange(a,b,h):
    puntos_x.append(x)
    k1=h*f(x,t)
    k2=h*f(x+0.5*k1,t+0.5*h)
    x=x+k2
    puntos_t.append(t)

plt.plot(puntos_t,puntos_x, color='skyblue', label= 'Metodo de Runge-kutta')
plt.grid()
plt.xlabel('t')
plt.legend()
plt.ylabel('x(t)')
plt.title('Metodo de Runge-kutta')
plt.show()
```
