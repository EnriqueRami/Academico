# **Metodo crudo de MonteCarlo**

```python
import random as random
import math as m
import numpy as np

def crudo(f,a,b,N):
  s_p=0
  for i in range(0,N):
    x=random.uniform(a,b)
    s_p+=f(x)
  return (b-a)*float(s_p/N)
```

```python
f= lambda x: x**4-2*x+1
print(crudo(f,0,2,10000))
```

```python
def g(x):
  return (np.exp**(-1*x))/(1+(x-1)**2)
xs=[float(i/50) for i in range(int(50*np.pi*2))]
ys=[f(x) for x in xs]
import matplotlib.pyplot as plt
plt.plot(xs,ys)
plt.show()
```

```python
print(crudo(g,0,10,1000))
```
