```python
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 50)


plt.plot(x, np.sin(x),linestyle = "--", color = "g", label = "Seno")
plt.plot(x, np.cos(x),linestyle = 'None', marker = 'v', color = "blue", label = "Coseno")
plt.plot(x, 0.01*np.tan(x), marker = '*', color = "orange", label = "Tangente")

# Etiquetas visuales

plt.legend()                           # Con esta instruccin mostramos las etiquetas de cada grfica
plt.grid()                             # Con esta otra agregamos un mallado gris clar
plt.title("Funciones trigonomtricas") # Agrega un ttulo
plt.xlabel("x")                        # Agrega una etiqueta al eje x
```

```python
fig, ax = plt.subplots(2,2, figsize = (12,8))

fig.suptitle("Dos grficas")

ax[0,0].plot(x, np.sin(x),'b')
ax[0,0].set_title("Grfica 1")


ax[0,1].plot(x, np.cos(x),'r')
ax[0,1].set_title("Grfica 2")

ax[1,0].plot(x, np.sin(x) + np.cos(x),'y')
ax[1,0].set_title("Grfica 3")

ax[1,1].plot(x, np.sin(x) - np.cos(x),'g')
ax[1,1].set_title("Grfica 4")


```

### Grficas Polares

```python
ax = plt.subplot(polar = True)  # ax es una matriz de nxm en la cual, cada entrada es una grfica
                          # fig es la figura completa
r = 4
k = 6
theta = np.linspace(0,2*np.pi,100)

ax.plot(theta, r*np.cos(k*theta), label = "Cos")
ax.plot(theta, r*np.sin(k*theta), label = "Sin", color = 'red')

ax.legend()
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, subplot_kw=dict(projection='polar'))

theta = np.linspace(0,2*np.pi,100)
r = 2

ax1.plot(theta, r*np.cos(2*theta))
ax2.plot(theta, r*np.cos(4*theta))
ax3.plot(theta, r*np.cos(6*theta))
ax4.plot(theta, r*np.cos(8*theta))
```

## Grficos 3D

```python

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

ax = plt.axes(projection ='3d')
```

```python
z = np.linspace(0,10,100)

y = z*np.sin(z)
x = z*np.cos(z)
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')

ax.plot3D(x,y,z, 'red')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')

ax.scatter(x,y,z, 'red')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')


c = x + y
ax.scatter(x,y,z, c = c)     # c = cmap
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.linspace(0,10,100)

x,y = np.meshgrid(x,y)

z = x**2 + y**2

fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection ='3d')
ax.plot_surface(x,y,z)
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x,y = np.meshgrid(x,y)

z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection ='3d')
ax.plot_surface(x,y,z, cmap = 'cool', alpha = 0.9)


ax.set_title('Grfica 3D')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x,y = np.meshgrid(x,y)

z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection ='3d')
ax.contour(x,y,z, cmap = 'cool')


ax.set_title('Grfica 3D')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
import plotly.graph_objects as go
```

```python
x = np.linspace(0,50,100)
y = np.linspace(0,50,100)

x_grid,y_grid = np.meshgrid(x, y)

z = np.cos(x_grid/2) + np.sin(y_grid/2)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])   # Aqu x,y son vectores pero z s es una matriz
fig.show()
```

```python
fig.update_layout(title=dict(text='Grfica 3D'),
                  autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
```

```python
fig.update_traces(contours_z=dict(show=True,
                                  usecolormap=True,
                                  highlightcolor="limegreen",
                                  project_z=True))
```

```python
x = np.linspace(0,50,100)
y = np.linspace(0,50,100)

x_grid,y_grid = np.meshgrid(x, y)

z = (x_grid + y_grid)/2

z1 = z + 10
z2 = z1 + 10

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y),
                      go.Surface(z=z1, x=x, y=y),
                      go.Surface(z=z2, x=x, y=y)])

fig.update_layout(title=dict(text='Grfica 3D'),
                  autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=2,
    specs=[[{'type': 'surface'}, {'type': 'surface'}],
           [{'type': 'surface'}, {'type': 'surface'}]])
```

```python
x = np.linspace(-5, 80, 10)
y = np.linspace(-5, 60, 10)
xGrid, yGrid = np.meshgrid(y, x)
z = xGrid ** 2 + yGrid ** 2
```

```python
fig.add_trace(
    go.Surface(x=x, y=y, z=z, colorscale='Viridis', showscale=False),
    row=1, col=1)   # Grfico superior izquierdo

fig.add_trace(
    go.Surface(x=x, y=y, z=z, colorscale='RdBu', showscale=False),
    row=1, col=2)   # Grfico superior derecho

fig.add_trace(
    go.Surface(x=x, y=y, z=z, colorscale='YlOrRd', showscale=False),
    row=2, col=1)   # Grfico inferior izquierdo

fig.add_trace(
    go.Surface(x=x, y=y, z=z, colorscale='YlGnBu', showscale=False),
    row=2, col=2)   # Grfico inferior derecho

fig.update_layout(
    title_text='3D subplots with different colorscales',
    height=800,
    width=800
)

fig.show()
```

# Graficas de Data Frames

```python
import plotly.express as px

df = px.data.iris()  # DataFrame propocionado por px


fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species")
fig.show()
```

```python
df = px.data.gapminder() # DataFrame propocionado por px

fig = px.scatter(df.query("year==2007"), x="gdpPercap", y="lifeExp", size="pop", color="continent",
           hover_name="country", log_x=True, size_max=60)
fig.show()
```

```python
df = px.data.gapminder()
fig = px.scatter_geo(df, locations="iso_alpha", color="continent", hover_name="country", size="pop",
               animation_frame="year", projection="natural earth")
fig.show()
```

```python
df = px.data.gapminder()
fig = px.choropleth(df, locations="iso_alpha", color="lifeExp", hover_name="country", animation_frame="year", range_color=[20,80])
fig.show()
```
