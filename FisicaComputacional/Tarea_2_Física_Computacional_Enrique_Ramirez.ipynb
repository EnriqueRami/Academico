```python
def promedio(datos):
    suma = 0
    for k in datos:
        suma += k

    if len(datos) == 0:
        return 0

    return suma / len(datos)
```

```python
def regresion_lineal(X, Y):

    x_p = promedio(X)
    y_p = promedio(Y)

    num = 0
    for x, y in zip(X, Y):
      num += (x - x_p) * (y - y_p)

    den = 0
    for x in X:
      den += (x - x_p) ** 2

    if den == 0:
        return "Error: La varianza de X es cero, no se puede calcular la regresión."

    if len(X) != len(Y):
        return "Error: Las listas X y Y deben tener la misma longitud."

    b1 = num / den
    b0 = y_p - b1 * x_p

    return b0, b1
```

```python
def coeficiente_r2(Y, F):

    rss=0
    for y, f in zip(Y, F):
        rss += (y - f) ** 2

    tss=0
    for y in Y:
        tss += (y - promedio(Y)) ** 2

    if tss == 0:
        return 0

    return 1 - rss / tss
```

##**Ejercicio 2**

```python
import random

def crear_mazos():
    baraja = []
    for valor in range(1, 14):
        baraja += [valor] * 4

    random.shuffle(baraja)

    mazo_A = baraja[:26]
    mazo_B = baraja[26:]

    return mazo_A, mazo_B

mazo_A, mazo_B = crear_mazos()
basura = []
ronda = 1

while len(mazo_A) > 0 and len(mazo_B) > 0:
    carta_A = mazo_A.pop(0)
    carta_B = mazo_B.pop(0)

    print(f"\nRonda {ronda}:")
    print("Cartas:", carta_A, "vs", carta_B)

    if carta_A > carta_B:
        mazo_A.append(carta_A)
        mazo_A.append(carta_B)
        print("Jugador A gana la ronda", ronda)

    elif carta_B > carta_A:
        mazo_B.append(carta_B)
        mazo_B.append(carta_A)
        print("Jugador B gana la ronda", ronda)

    else:
        basura.append(carta_A)
        basura.append(carta_B)
        print("Empate en la ronda", ronda, ". Cartas desechadas:", carta_A, "y", carta_B)

    ronda += 1

print("\n Resultado Final")
if len(mazo_A) == 0 and len(mazo_B) == 0:
    print("¡Empate! Ambos jugadores se quedaron sin cartas.")

elif len(mazo_A) == 0:
    print("¡Jugador B ha ganado!")
    print("Mazo de B:", mazo_B)

else:
    print("¡Jugador A ha ganado!")
    print("Mazo de A:", mazo_A)

print("Mazo basura:", basura)
```

# **Ejercicio 3**

```python
import numpy as np

def transpuesta(matriz):
    return np.transpose(matriz)

def determinante_2x2(matriz):
    return matriz[0][0] * matriz[1][1] - matriz[0][1] * matriz[1][0]

def matriz_cofactores(matriz):
    n = len(matriz)
    cof = np.zeros((n, n))

    for i in range(n):
        for j in range(n):

            submatriz = [fila[:j] + fila[j+1:] for fila in (matriz[:i] + matriz[i+1:])]

            cof[i][j] = ((-1) ** (i + j)) * determinante_2x2(submatriz)

    return cof

def determinante_3x3(matriz):
    a11, a12, a13 = matriz[0]
    a21, a22, a23 = matriz[1]
    a31, a32, a33 = matriz[2]

    det = (a11 * (a22 * a33 - a23 * a32) -
           a12 * (a21 * a33 - a23 * a31) +
           a13 * (a21 * a32 - a22 * a31))

    return det

def inversa_matriz_3x3(matriz):
    det = determinante_3x3(matriz)

    if det == 0:
        return "La matriz no es invertible (su determinante es cero)."

    cof = matriz_cofactores(matriz)
    adj = transpuesta(cof)
    inversa = (1 / det) * adj

    return inversa

matriz = [[1, 2], [3, 4]]

transpuesta_matriz = transpuesta(matriz)
print("Transpuesta de la matriz:")
print(transpuesta_matriz)

determinante = determinante_2x2(matriz)
print("Determinante de la matriz:")
print(determinante)

A = [[4, 7, 2],
     [3, 5, 1],
     [2, 6, 8]]

print("\nMatriz de cofactores:")
print(matriz_cofactores(A))

det_A = determinante_3x3(A)
print("\nEl determinante de A es:", det_A)

A_inv = inversa_matriz_3x3(A)
print("\nLa inversa de A es:")
print(A_inv)
```

# **Ejercicio 5**

```python
def factorial(n):
        fact = 1
        for i in range(1, n + 1):
            fact= fact * i
        return fact

from scipy.special import eval_hermite
import numpy as np
def psi_n(n, x, m=1, omega=1, hbar=1):
    norm = np.sqrt(1 / (2**n * factorial(n))) * (m * omega / (np.pi * hbar))**(1/4)
    xi = np.sqrt(m * omega / hbar) * x
    return norm * np.exp(-m * omega * x**2 / (2 * hbar)) * eval_hermite(n, xi)

import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
plt.figure(figsize=(10, 8))
colors = plt.cm.viridis(np.linspace(0, 1, 8))


for n in range(8):
    y = psi_n(n, x) + n
    plt.plot(x, y, color=colors[n])
    plt.fill_between(x, n, y, color=colors[n], alpha=0.4)

V = 0.5 * x**2
plt.plot(x, V, 'k--')

for n in range(8):
    plt.axhline(y=n, color='gray', linestyle='--', linewidth=0.8)

plt.xlabel("x")
plt.ylabel("n")
plt.title("Funciones de onda del oscilador armónico")
plt.xlim(-4, 4)
plt.ylim(0, 8)
plt.show()

```

# **Ejercicio 6**

```python
import random
import matplotlib.pyplot as plt

def suma_dados():
    return random.randint(1, 6) + random.randint(1, 6)

def simulamos_lanzamientos(N):
    resultados = []
    for _ in range(N):
        resultados.append(suma_dados())
    return resultados


def calculamos_probabilidades(resultados):
    probabilidades = {}
    for i in range(2, 13):
        probabilidades[i] = resultados.count(i) / len(resultados)
    return probabilidades


def graficar_distribucion(probabilidades):
    numeros = list(probabilidades.keys())
    probs = list(probabilidades.values())

    plt.bar(numeros, probs, color='blue')
    plt.xlabel('Suma de los dados')
    plt.ylabel('Probabilidad')
    plt.title('Distribución de probabilidades de la suma de dos dados')
    plt.xticks(numeros)
    plt.show()


N = 10000


resultados = simulamos_lanzamientos(N)


probabilidades = calculamos_probabilidades(resultados)


graficar_distribucion(probabilidades)
```
